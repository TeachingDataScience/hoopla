```python
from mpl_toolkits.mplot3d import Axes3D

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
#import matplotlib.pyplot as plt






import seaborn as sns
```

```python
df = pd.read_csv("../nba-player-tracking/source/players.csv")
```

```python
# Find columns with nulls greater than 5
nullcols = df.T[df.isnull().sum() > 5].index
```

```python
# Remove extra columns
df = df.T[~df.columns.isin(nullcols)].T
```

```python
# Choose variables
cols = df.columns[18:]
playframe = df[cols]
```

```python
# Remove NA, preliminary hack. #TODO interpolate later
playframe = playframe.fillna(0)
```

```python
# choose clusters to try
k = list(range(2, 11))
scores = dict.fromkeys(k)
```

```python
# Kmeans

data = scale(playframe.values)
```

```python
# Get silhouette
for size in k: 
    kmeans = KMeans(n_clusters=size)
    kmeans.fit(data)
    scores[size] = silhouette_score(data, kmeans.labels_)
```

```python
# Plot silhouette
plt.figure(figsize(10,5))
plt.plot(k, scores.values())
plt.title('K-means clustering of NBA Players, 2014')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
```

```python
# Kmeans, choose clusters
kmeans = KMeans(n_clusters=7)
kmeans.fit(data)
clusters = pd.DataFrame(kmeans.cluster_centers_, columns=playframe.columns)
```

```python
# Add cluster assignments to initial frame
playframe['cluster'] = kmeans.labels_
```

## Review Clusters and Assignments

```python
#Create cluster scores frame
clusterscores = pd.DataFrame(kmeans.cluster_centers_,columns=cols)
```

```python
clusterscores = dropzeros(clusterscores)
```

```python
graph_metrics(clusterscores,5)
```

## Cluster Histograms

```python
# Add cluster label
df['cluster'] = kmeans.labels_
```

```python
plt.figure(figsize=(8,20))

plt.suptitle('Points Scored by Cluster',fontsize=14)
for c in range(0,kmeans.n_clusters):
    plt.subplot(7, 1, c+1)

    df[df.cluster == c].pts_48.hist(alpha=.4,range=(0,25),bins=15)
    plt.xlabel('Points Scored per 48 Minutes')
    plt.title("Group " + str(c))

#plt.show()
```

### Histogram Highlight Combo Charts

```python
plt.figure(figsize=(20,40))

plt.suptitle('Cluster Details',fontsize=14)

clusternum = kmeans.n_clusters

for c in range(0,clusternum):
    print c
    
    # Top Features
    
    varnum = 5 # set varnum
    plt.subplot(clusternum+2, 3, c+1)
    cc = clusterscores.ix[c]
    c_abs = cc.abs()
    c_abs.sort(ascending=False)
    
    csub = cc[c_abs.index[0:varnum]]
    csub.sort(ascending=True,inplace=True)
    csub.plot(kind='barh')
    title = 'Cluster ' + str(c) + " defining metrics"
    plt.title(title)
    
    
    
    
    # Histogram
    plt.subplot(clusternum+2, 3, c+2)

    df[df.cluster == c].pts_48.hist(alpha=.4,range=(0,25),bins=15)
    plt.xlabel('Points Scored per 48 Minutes')
    plt.title("Group " + str(c))

plt.show()
```

```python
    
def dropzeros(s):
    """
    drop zeros from a series
    """
    s = s.T[s.max() != 0]
    return(s.T)
```
