# Try basic clustering

Based first on [treycausey](https://github.com/treycausey/thespread/blob/master/clustering_wrs.py)

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import scale
```

```python
print df.columns[18:]
```

```python
# Iterate through [2, 10] to find k, the number of clusters
k = list(range(2, 11))
scores = dict.fromkeys(k)
```

```python
cols = df.columns[18:]
```

```python
playframe = df[df.columns[18:]]
```

```python
playframe = playframe.fillna(0)
```

```python
data = scale(playframe.values)
```

```python
for size in k:
    kmeans = KMeans(n_clusters=size)
    kmeans.fit(data)
    scores[size] = silhouette_score(data, kmeans.labels_)
```

```python
%pylab inline
```

```python
plt.plot(k, scores.values())
plt.title('K-means clustering of NBA Players, 2014')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
```

```python
clusters = pd.DataFrame(kmeans.cluster_centers_, columns=playframe.columns)
```

```python
kmeans.n_clusters
```

```python
playframe['cluster'] = kmeans.labels_
```

```python
df['cluster'] = kmeans.labels_
```

### Make Frame with Variables and Cluster Centers

```python
qscores = pd.DataFrame(kmeans.cluster_centers_,columns=cols)
```

```python
qscores = dropzeros(qscores)
```

```python

qscores.T.plot(kind='barh',figsize=(5,20))
plt.title('Cluster Scores by Metric')
```

## Evaluate Top Categories per Cluster

```python
def graph_metrics(frame,varnum):
    """
    Create graph of absolute top scoring variables in frame
    """

    for i in range(0,kmeans.n_clusters):
        print i
        c = frame.ix[i]
        c_abs = c.abs()
        c_abs.sort(ascending=False)
        plt.figure(figsize(5,varnum/5))
        title = 'Cluster ' + str(i) + " defining metrics"
        plt.title(title)
        csub = c[c_abs.index[0:varnum]]
        csub.sort(ascending=True,inplace=True)
        csub.plot(kind='barh')

        
def dropzeros(s):
    """
    drop zeros from a series
    """
    s = s.T[s.max() != 0]
    return(s.T)
```

```python
qscores
```

```python
graph_metrics(qscores,40)
```

```python
scores_T = qscores.T.sort([0,1,2],ascending=False)
scores_T.plot(kind='barh',figsize=(10,20))
plt.title('Cluster Scores by Metric')
```

```python
scores_T = qscores.T[[1,2]].sort([1,2],ascending=True)
scores_T.plot(kind='barh',figsize=(10,20),colors=['g','r'])
plt.title('Cluster Scores by Metric')

```
